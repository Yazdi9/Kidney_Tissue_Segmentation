# Raw Interpolated

````python
import os
import nibabel as nib

dset_dir = r"C:\Users\jchen\Desktop\Datasets\kits19\data"
i = 14
case = f"case_000{i}"
image = nib.load(os.path.join(dset_dir, case, "imaging.nii.gz")).get_fdata().squeeze()
mask = nib.load(os.path.join(dset_dir, case, "segmentation.nii.gz")).get_fdata().squeeze()

print(f"Case: {case}; Shape: {image.shape}")
assert mask.shape == image.shape
````

````python
print(nib.load(os.path.join(dset_dir, case, "imaging.nii.gz")).affine)
````

````python
import ipyvolume as ipv
ipv.quickvolshow(mask)
````
